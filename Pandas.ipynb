## merge()

![](merge_func.png)

```python
df_1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35]
} )
df_1
```

```python
df_2 = pd.DataFrame({
    "city": ["new york","chicago","orlando" , 'india'],
    "humidity": [65, 68 , 75 , 86] , 
})
df_2
```

```python
# inner join by default
# based on 'city'
df3 = pd.merge(df_1 , df_2 , on='city') 
df3
```

```python
# outer join based on 'city'

df4 = pd.merge(df_1 , df_2 , on='city' , how = 'outer')
df4
```

```python
# left join based on'city'

df5 = pd.merge(df_1 , df_2 , on='city' , how = 'left')
df5
```

```python
# right join based on'city'
# Indicator indicates the presence of rows in dataframes

df6 = pd.merge(df_1 , df_2 , on='city' , how = 'right',indicator=True)
df6
```

```python
df_1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "windspedd":[1,2,3]
} )


df_2 = pd.DataFrame({
    "city": ["new york","chicago","orlando" , 'india'],
    "temperature": [21,66,77 , 88] , 
    "humidity": [65, 68 , 75 , 86]  
})
```

```python
df_1
```

```python
df_2
```
