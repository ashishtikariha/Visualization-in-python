## Iterating in numpy

```python
## np.genformtxt()  ---> Load data from text file
## .csv ---> comma seperated value ---> data are seperated by comma ---> delimiter =','
## .tsv ---> tab seperated value ---> data are seperated by tab ---> delimiter ='    '

## usecols=1 ---> Read only column at index 1
## delimiter = ','   ---> Telling numpy that data are seperated by ','
## skip_header = 1   ---> The number of lines to skip at the beginning of the file

population_data = np.genfromtxt('population_by_country_2020.csv' , usecols = 1 , delimiter = ',' , skip_header = 1)
print(population_data)      # All the elements on 2nd column will be printed
```

```python
print(population_data.max())
print(population_data.min())
print(population_data.mean())
print(population_data.sum())
```

```python
print(population_data.shape)
print(population_data.ndim)
print(population_data.size)
```

## copy()
## view()

```python
#copy ----> It creates new array with new memory location
# Changes in new array doesnt reflect to old array

arr = np.array([1,2,3,4,5])
x = arr.copy()
x[0] = 200

print(arr)
print(x)
```

```python
# view ----> It creates new array that points to the memory location of old array
# Changes in new array reflected to old array

arr = np.array([1,2,3,4,5])
x = arr.view()
x[0] = 200

print(arr)
print(x)
```

## How computer process image ?

## For computer image is nothing but set of pixels which are stored in array

#### For black & white image...The pixels are stored in 2-D array
#### When we say 400*200 image...it means image is divided into 400 rows & 200 columns
#### each element posses some pixel from 0-255...each corrosponds to different color shade

#### For color image...The pixels are stored in 3-D array
#### When we say 400**200**200 image...it means image is divided into 400 rows , 200 columns , 200 depth
#### each element posses some pixel from 0-255...each corrosponds to different color shade

```python
# loading the images from skimage database

from skimage import data         # importing all data stored in scikit-image library

astronaut = data.astronaut()     # an astronaut image is stored in astronaut()
print(astronaut)                 # see a colour image is stored in 3-D array

# see each elements(pixels) are in range 0-255
```
